```python
import os 
import re
import pandas as pd
import pickle as pk
import numpy as np
import formate_patterns
import extract_features

def load_df(path_df, sep="\t", encoding="utf-8"):
    return pd.read_csv(path_df, sep=sep, encoding=encoding)

def load_pk(path_pk):
    with open(path_pk, "rb") as input:
        return pk.load(input)

def dump_pk(obj_to_save, path_pk):
    with open(path_pk, "wb") as ouput:
        return pk.dump(obj_to_save, ouput)
```
